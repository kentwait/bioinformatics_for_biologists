```python
# Generate fake reads from the original sequencea
import random

def fake_reads(original_seq, read_len, coverage=30):
    """Generate reads from a template sequence.
    
    Parameters
    ----------
    original_seq : str
    read_len : int
        Length of simulated read
    coverage : int
        Number of times to repeated chop the template sequence to generate reads
        
    Returns
    -------
    list
        List of simulated reads with a uniform read length
    """
    reads = []
    for _ in range(coverage):
        rand_start = random.randint(0, read_len-1)
        reads += [original_seq[start:start+read_len] 
                  for start in range(rand_start, len(original_seq), read_len)
                  if start+read_len <= len(original_seq)
                 ]
    return reads
```

```python
# Find if there is an overlap between a pair of sequences
def suffixPrefixMatch(x, y, k):
    """Return length of longest suffix of x of length at least k that
    matches a prefix of y.  Return 0 if there no suffix/prefix
    match has length at least k.
    
    Parameters
    ----------
    x, y : str
        sequence reads
    k : int
        Minimum number of overlapping nucleotides. 
        These are the set of nucleotides found at the end of x that is also found in y
        
    Returns
    -------
    int
        Returns 0 if no match was found with the specified minimum length k.
        Otherwise, returns the length of the remaining prefix.
    """
    if len(x) < k or len(y) < k:
        return 0
    idx = len(y) # start at the right end of y
    # Search right-to-left in y for length-k suffix of x
    while True:
        hit = y.rfind(x[-k:], 0, idx)
        if hit == -1: # not found
            return 0
        ln = hit + k
        # See if match can be extended to include entire prefix of y
        if x[-ln:] == y[:ln]:
            return ln # return length of prefix
        idx = hit + k - 1 # keep searching to left in Y
    return -1
```

```python
# Test
read1 = "GAGCCTGGGAGCTCTCTGGCTAACTAGGGAACC"
min_overlap_len = 3
matches = []
num_matches = 0
for read in reads[1:]:
    match_len = suffixPrefixMatch(read1, read, min_overlap_len)
    matches.append((read, match_len))
    if match_len > 0:
        num_matches += 1
print(num_matches)
```

```python
min_overlap_len = 3

raw_reads = list(reads)
c = 0
while len(raw_reads) > 1:
    matches = []
    for i, read1 in enumerate(raw_reads):
        for j, read2 in enumerate(raw_reads):
            if i < j:
                match_len = suffixPrefixMatch(read1, read2, min_overlap_len)
                if match_len > 0:
                    matches.append((i, j, match_len))
                else:
                    match_len = suffixPrefixMatch(read2, read1, min_overlap_len)
                    matches.append((j, i, match_len))
    i1, i2, match_len = sorted(matches, key=lambda x: x[2], reverse=True)[0]
    if match_len == 0:
        break
    
    if i1 < i2:
        read1 = raw_reads.pop(i1)
        read2 = raw_reads.pop(i2-1)
    else:
        read2 = raw_reads.pop(i2)
        read1 = raw_reads.pop(i1-1)
    
    if match_len == len(read2):
        merged_read = read2
    else:
        merged_read = read1[:-match_len] + read2
    raw_reads.insert(0, merged_read)
    # Print stats
    offset_read2 = ' '*len(read1[:-match_len]) + read2
    print("Number of substrings:", len(raw_reads))
    print("Match length:", match_len)
    for i in range(80, len(merged_read), 80):
        print("r1:    ", read1[i-80:i])
        print("r2:    ", offset_read2[i-80:i])
        print("merged:", merged_read[i-80:i])
        print('')
    print("r1:    ", read1[i:] if len(read1) > 80 else read1)
    print("r2:    ", offset_read2[i:] if len(offset_read2) > 80 else offset_read2)
    print("merged:", merged_read[i:] if len(merged_read) > 80 else merged_read)
        
    print("Iterations:", c)
    print("-"*80)
    
    c += 1
```
